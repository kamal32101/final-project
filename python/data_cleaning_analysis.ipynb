```python
import pandas as pd

mat = pd.read_excel(r"C:\Users\DELL\OneDrive\Desktop\data\mat.xlsx")
por = pd.read_excel(r"C:\Users\DELL\OneDrive\Desktop\data\por.xlsx")

# نضيف عمود يحدد المادة
mat["subject"] = "math"
por["subject"] = "portuguese"

# ندمج
df = pd.concat([mat, por], ignore_index=True)
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
df["avg_grade"] = ((df["G1"] + df["G2"] + df["G3"]) / 3).round(2)
df.head()
```

```python
df["status"] = df["G3"].apply(lambda x: "Pass" if x >= 10 else "Fail")
def level(x):
    if x >= 15:
        return "Excellent"
    elif x >= 12:
        return "Good"
    elif x >= 10:
        return "Pass"
    else:
        return "Fail"

df["level"] = df["G3"].apply(level)
df.head()
```

```python
df = df.rename(columns={
    "sex":"gender",
    "famsize":"family_size",
    "Pstatus":"parent_status",
    "Medu": "mother_education",
    "Fedu": "father_education",
    "Mjob":"mother_job",
    "Fjob":"father_job",
    "traveltime":"travel_time",
    "studytime":"study_time",
    "schoolsup":"school_support",
    "famsup":"family_support",
    "higher":"wants_higher_education",
    "famrel":"family_relationship_quality",
    "goout":"go_out",
    "G1": "first_grade",
    "G2": "second_grade",
    "G3": "final_grade",
    "absences": "absence_days"
})
```

```python
df = df.rename(columns={
    "parent_status":"parents_status"
})
```

```python
df = df.drop(columns=["Walc", "Dalc"])
```

```python
df.columns
```

```python
df["gender"] = df["gender"].map({
    "M": "Male",
    "F": "Female"
})
df["school"] = df["school"].map({
    "GP": "Gabriel Pereira",
    "MS": "Mousinho da Silveira"
})
df["address"] = df["address"].map({
    "U": "Urban",
    "R": "Rural"
})
df["family_size"] = df["family_size"].map({
    "LE3": "Small Family",
    "GT3": "Large Family"
})
df["parents_status"] = df["parents_status"].map({
    "T": "Together",
    "A": "Apart"
})

df.head()
```

```python
education_map = {
    0: "None",
    1: "Primary (up to 4th)",
    2: "5th to 9th grade",
    3: "Secondary education",
    4: "Higher education"
}
traveltime_map = {
    1: "less than 15 min",
    2: "15 to 30 min",
    3: "30 to 60 min",
    4: "greater than 60 min"
}
studytime_map = {
    1: "less than 2 hours",
    2: "2 to 5 hours",
    3: "5 to 10 hours",
    4: "greater than 10 hours"
}
famrel_map = {
    1: "Very bad",
    2: "Bad",
    3: "Average",
    4: "Good",
    5: "Excellent"
}
freetime_map = {
    1: "Very low",
    2: "Low",
    3: "Medium",
    4: "High",
    5: "Very high"
}
goout_map = {
    1: "Very low",
    2: "Low",
    3: "Medium",
    4: "High",
    5: "Very high"
}
health_map = {
    1: "Very bad",
    2: "Bad",
    3: "Average",
    4: "Good",
    5: "Very good"
}
```

```python
df["mother_education"] = df["mother_education"].map(education_map)
df["father_education"] = df["father_education"].map(education_map)

df["travel_time"] = df["travel_time"].map(traveltime_map)
df["study_time"] = df["study_time"].map(studytime_map)

df["family_relationship_quality"] = df["family_relationship_quality"].map(famrel_map)
df["freetime"] = df["freetime"].map(freetime_map)
df["go_out"] = df["go_out"].map(goout_map)

df["health"] = df["health"].map(health_map)
```

```python
cols = [
    "school",
    "gender",
    "address",
    "family_size",
    "parents_status",
    "mother_education",
    "father_education",
    "mother_job",
    "father_job",
    "reason",
    "guardian",
    "travel_time",
    "study_time",
    "school_support",
    "family_support",
    "paid",
    "activities",
    "nursery",
    "wants_higher_education",
    "internet",
    "romantic",
    "family_relationship_quality",
    "freetime",
    "go_out",
    "health",
    "status",
    "level",
    "subject"
]
```

```python
for col in cols:
    df[col] = df[col].astype("category")
```
